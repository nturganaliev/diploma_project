```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import requests

from datetime import datetime
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# Binance API endpoint for historical data
api_url = 'https://api.binance.com/api/v3/klines'

# Parameters for the API request (Bitcoin, 1 day interval, 1000 data points)
params = {
    'symbol': 'BTCUSDT',
    'interval': '1d',
    'limit': 10000
}

# Send GET request to Binance API
response = requests.get(api_url, params=params)
data = response.json()



# Extracting relevant data from the API response
df = pd.DataFrame(data, columns=['openTime', 'open', 'high', 'low', 'close', 'volume', 'closeTime', 'quoteAssetVolume', 'numberOfTrades', 'takerBuyBaseAssetVolume', 'takerBuyQuoteAssetVolume', 'ignore'])

# Renaming columns for clarity
df.columns = ['Open Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close Time', 'Quote Asset Volume', 'Number of Trades', 'Taker Buy Base Asset Volume', 'Taker Buy Quote Asset Volume', 'Ignore']


df.drop(['Close Time', 'Quote Asset Volume', 'Number of Trades', 'Taker Buy Base Asset Volume', 'Taker Buy Quote Asset Volume', 'Ignore'], axis=1, inplace=True)

# Converting timestamps to datetime format
df['Date'] = pd.to_datetime(df['Open Time'], unit='ms')

# Feature engineering (you may need to add more features based on your data)
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year



# if df is not None:
#     # Save DataFrame to CSV file
#     df.to_csv('bitcoin_prices.csv', index=False)
#     print("DataFrame saved to 'bitcoin_prices.csv'")
# else:
#     print("Failed to fetch Bitcoin price data.")

cols = df.columns.drop('Open Time')
cols
df[cols] = df[cols].apply(pd.to_numeric)

# Initialize and train the Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
X = df[['Open', 'High', 'Low', 'Volume']]
y = df['Close']
model.fit(X,y)

# Generate predictions for the entire dataset
predicted_prices = model.predict(X)

# Plot actual vs predicted
plt.figure(figsize=(18, 6))
plt.title('Actual vs Predicted Bitcoin Price')
plt.plot(df['Open Time'], df['Close'], label='Actual Price')
plt.ylabel('Price')
plt.plot(df['Open Time'], predicted_prices, label='Predicted Price', color='orange')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()








# # Calculate Mean Absolute Error (MAE)
# mae = mean_absolute_error(y, predicted_prices)

# # Calculate Mean Squared Error (MSE)
# mse = mean_squared_error(y, predicted_prices)

# # Calculate R-squared (R2) score
# r2 = r2_score(y, predicted_prices)

# # Calculate Root Mean Squared Error (RMSE)
# rmse = math.sqrt(mse)

# print(f"Mean Absolute Error (MAE): {mae}")
# print(f"Mean Squared Error (MSE): {mse}")
# print(f"Root Mean Squared Error (RMSE): {rmse}")
# print(f"R-squared (R2) Score: {r2}")


# # Convert predicted_prices to numpy array and cast to float for compatibility
# predicted_prices_np = np.array(predicted_prices, dtype=float)

# # Calculate TWAP for actual prices
# actual_prices = np.array(y, dtype=float)
# actual_twap = np.mean(actual_prices)

# # Calculate TWAP for predicted prices
# predicted_prices_twap = np.mean(predicted_prices_np)

# # Calculate the accuracy using TWAP
# accuracy_twap = 100 * (1 - abs(actual_twap - predicted_prices_twap) / actual_twap)

# print(f"TWAP of Actual Prices: {actual_twap}")
# print(f"TWAP of Predicted Prices: {predicted_prices_twap}")
# print(f"Accuracy using TWAP: {accuracy_twap:.2f}%")

# # Convert price and volume data to numerical format
# actual_prices = np.array(y, dtype=float)
# actual_volume = np.array(df['Volume'], dtype=float)

# # Calculate VWAP for actual prices
# actual_vwap = np.sum(actual_prices * actual_volume) / np.sum(actual_volume)

# # Convert predicted prices to numerical format
# predicted_prices = np.array(predicted_prices, dtype=float)

# # Calculate VWAP for predicted prices
# predicted_prices_vwap = np.sum(predicted_prices * actual_volume) / np.sum(actual_volume)

# # Calculate the accuracy using VWAP
# accuracy_vwap = 100 * (1 - abs(actual_vwap - predicted_prices_vwap) / actual_vwap)

# print(f"VWAP of Actual Prices: {actual_vwap}")
# print(f"VWAP of Predicted Prices: {predicted_prices_vwap}")
# print(f"Accuracy using VWAP: {accuracy_vwap:.2f}%")

df.head()
```
